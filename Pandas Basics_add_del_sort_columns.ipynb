```python
import numpy as np
import pandas as pd
```

```python
#add header as col_Names
col_Names=["Language1", "Language2", "Right", "Wrong"]
df=pd.read_csv('file.csv',names=col_Names)
```

```python
df
```

```python
df2 = df.copy()
```

```python
#add new column which is sum of right+wrong
df2["Total counts"]=df2["Right"]+df2["Wrong"]
```

```python
df2
```

```python
#add new column which is sum of right+wrong
df2["Right %"]=df2["Right"]*100/df2["Total counts"]
df2["Wrong %"]=df2["Wrong"]*100/df2["Total counts"]
```

```python
df2
```

```python
#add new column with "0"s
df2["F"]=0
```

```python
df2
```

```python
#delete column 
df2.drop(columns=["F"])
```

```python
#new df without "F"
df3 = df2.drop(columns=["F"])
```
